```python
df = pd.read_csv('/Users/mairagutierrez/Documents/Ironhack/PROJECTS/project-IV--STREAMING-SERVICES-ANALYSIS/data/streaming_services_df.csv')

def process_genre(genre):
    if pd.isna(genre):
        return None  # Return None or a default value if the genre is NaN
    genre = genre.strip()
    genres = genre.split(',')
    # Check for 'Kids''
    for g in genres:
        if "Kids'" in g:
            return g.strip()
    # Handle cases with only one or no genre
    if len(genres) == 0:
        return None  # Return None or a default value if there are no genres
    elif len(genres) == 1:
        return genres[0].strip()
    else:
        # Skip 'International' if it's the first genre and there are at least two genres
        return genres[1].strip() if "International" in genres[0] else genres[0].strip()

# Apply the function to the 'listed_in' column
df['listed_in'] = df['listed_in'].apply(process_genre)
df
```
