### Total Titles by Streaming Service:

```python
query1 = "SELECT streaming_service, COUNT(*) as total_titles FROM streaming_services GROUP BY streaming_service;"
pd.read_sql_query(query1, engine)
```

### Average Release Year and Added Year by Streaming Service:

```python
query2 = """
SELECT 
    streaming_service, 
    CAST(AVG(release_year) AS UNSIGNED) as avg_release_year,
    CAST(AVG(CASE WHEN streaming_service IN ('Netflix', 'Disney') THEN year ELSE NULL END) AS UNSIGNED) as avg_added_year
FROM streaming_services
GROUP BY streaming_service;
"""
pd.read_sql_query(query2, engine)
```

### Most Common Genre for Each Streaming Service:

```python
query3 = "SELECT streaming_service, listed_in, COUNT(*) as genre_count FROM streaming_services GROUP BY streaming_service, listed_in ORDER BY streaming_service, genre_count DESC;"
pd.read_sql_query(query3, engine)
```

### Distribution of Movie Durations:

```python
query4 = "SELECT duration, COUNT(*) as count FROM streaming_services WHERE type = 'Movie' GROUP BY duration;"
pd.read_sql_query(query4, engine)
```

### Count of TV Shows by Country:

```python
query5 = "SELECT country, COUNT(*) as count FROM streaming_services WHERE type = 'TV Show' GROUP BY country;"
pd.read_sql_query(query5, engine)
```

### Rating Distribution:

```python
query6 = "SELECT rating, COUNT(*) as count FROM streaming_services GROUP BY rating;"
pd.read_sql_query(query6, engine)
```

### TOP 5 Most Frequent Directors for Each Streaming Service:

```python
query7 = """
SELECT streaming_service, director, total_titles
FROM (
    SELECT streaming_service, director, COUNT(*) as total_titles,
           ROW_NUMBER() OVER (PARTITION BY streaming_service ORDER BY COUNT(*) DESC) as row_num
    FROM streaming_services
    WHERE director IS NOT NULL
    GROUP BY streaming_service, director
) as ranked_directors
WHERE row_num <= 5;
"""

# Execute query and store results in a DataFrame
pd.read_sql_query(query7, engine)
```
